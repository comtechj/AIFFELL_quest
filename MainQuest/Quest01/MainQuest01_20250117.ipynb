```python
from Account import Account
from TransactionType import TransactionType
import re

account_list = []
def input_from_user():
    try:
        return int(input("초기 입금액을 입력하세요 : "))
    except ValueError:
        print("숫자를 입력하세요")
        return input_from_user()

def input_for_menu():
    try:
        return int(input("초기 입금액을 입력하세요 : "))
    except ValueError:
        print("숫자를 입력하세요")
        return input_from_user()

def show_choose_menu():
    for key, value in menu_dict.items():
        print(value['description'])
    try:
        menu_input = int(input('메뉴를 선택하세요: '))
    except ValueError:
        print('숫자를 입력하세요')
        return show_choose_menu()
    else:
        if menu_input not in menu_dict.keys():
            print('메뉴를 다시 선택하세요')
            return show_choose_menu()

    return menu_input

def input_account_number():
    account_number = input('계좌번호: ')
    if account_number is None or not re.match(r'\d{3}-\d{2}-\d{7}', account_number):
        print('계좌번호 형식이 올바르지 않습니다.')
        return input_account_number()
    return account_number

def create_new_account():
    name = input('이름: ')
    balance = input_from_user()
    account = Account(name, balance)
    account_list.append(account)
    print(f'계좌가 생성되었습니다. {account.display_info()}')
    return account

def find_account(account_number):
    filtered_accounts = filter(lambda acc: acc.account_number == account_number, account_list)
    if filtered_accounts is None:
        print('계좌번호가 없습니다.')
        return None
    elif len(list(filtered_accounts)) > 1:
        print('계좌번호가 중복됩니다.')
        return None
    else:
        return list(filtered_accounts)[0]

def deposit():
    account_number = input_account_number()
    account = find_account(account_number)
    if account is None:
        return False
    amount = input_from_user()
    account.deposit(amount)
    return True

def withdraw():
    account_number = input_account_number()
    account = find_account(account_number)
    if account is None:
        return False
    amount = input_from_user()
    account.withdraw(amount)
    return True

def print_transaction_history():
    account_number = input_account_number()
    account = find_account(account_number)
    if account is None:
        return False
    for index,(tr_type, amount, balance) in enumerate(account.get_transaction_history()):
        print(f"{index+1}회: {tr_type}, 금액: {amount}, 잔액: {balance}")
    return True

def print_account_info():
    account_number = input_account_number()
    account = find_account(account_number)
    if account is None:
        return False
    print(account)
    return True

def print_accounts():
    for account in account_list:
        print(account)

menu_dict = {
    1: {'menu': create_new_account , 'description': '1. 계좌 생성'},
    2: {'menu': deposit, 'description': '2. 입금'},
    3: {'menu': withdraw, 'description': '3. 출금'},
    4: {'menu': print_account_info, 'description': '4. 계좌 정보 출력'},
    5: {'menu': print_transaction_history, 'description': '5. 거래 내역 출력'},
    6: {'menu': exit, 'description': '6. 종료'}}


if __name__ == '__main__':
    while True:
        menu = show_choose_menu()
        if menu == 6:
            break
        menu_dict[menu]['menu']()

    print('프로그램 종료')
```

```python
import datetime as dt
class Transaction:
    def __init__(self, type, amount, balance):
        self.type = type
        self.amount = amount
        self.balance = balance
        self.date = dt.datetime.now()

    def __str__(self):
        return f'날짜: {self.date}, 내용: {self.type}, 금액: {self.amount}, 잔액: {self.balance}'
```

```python
from enum import Enum
class TransactionType(Enum):
    DEPOSIT = (1, '입금')
    WITHDRAW = (2, '출금')
    INCREASE_DEPOSIT_INTEREST = (3, '이자지급')

    def __init__(self, code, title):
        self.code = code
        self.title = title

    def __str__(self):
        return self.title
```
